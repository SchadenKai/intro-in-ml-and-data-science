```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../data/spotify_data.csv')
dataset.head()
```

```python
first_200 = dataset.iloc[:200]

less_than_1970 = first_200.loc[first_200['released_year'] < 1970]
less_than_1970
```

```python
greater_than_1982_and_less_than_1985 = first_200.loc[(first_200['released_year'] < 1985) & (first_200['released_year'] > 1982)]
greater_than_1982_and_less_than_1985
```

```python
greater_than_2012_and_less_than_2015 = first_200.loc[(first_200['released_year'] < 2015) & (first_200['released_year'] > 2012)]
greater_than_2012_and_less_than_2015
```

```python
greater_than_2019_and_less_than_or_equal_2022 = first_200.loc[(first_200['released_year'] <= 2022) & (first_200['released_year'] > 2019)]
greater_than_2019_and_less_than_or_equal_2022.head()
```
