```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Features
aguero = [32, 29, 33, 30, 32]
neymar = [39, 31, 20, 28, 23]
seasons = ['14/15', '15/16', '16/17', '17/18', '18/19']

aguero_mean = np.mean(aguero)
neymar_mean = np.mean(neymar)
```

```python
plt.plot(seasons, aguero, color='#ffa21f', label='Aguero', marker='o')
plt.plot(seasons, neymar, label='Neymar', marker='o')

plt.axhline(y=aguero_mean, color='#ffa21f', linestyle='--')
plt.axhline(y=neymar_mean, linestyle='--')

plt.title('Comparison of Aguero and Neymar goals per season')
plt.xlabel('Season')
plt.ylabel('Goals')
plt.legend()
```

```python
aguero_std = np.std(aguero)
neymar_std = np.std(neymar)

print("Aguero standard deviation: " + str(aguero_std))
print("Neymar standard deviation: " + str(neymar_std))
print("\nVariance:")
print("Aguero: " + str(np.var(aguero)))
print("Neymar: " + str(np.var(neymar)))
```

This means that Neymar has more variability on its scoring points between seasons, which means it is less consistent. Having lesser standard deviation value means that the player is more consistent. Since Aguero has less standard deviation, he is more consistent than Neymar.
